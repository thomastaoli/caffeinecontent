```python
import json
```

```python
with open('menu.json', 'r') as file:
    data = json.load(file)
```

```python
# Initialize an empty list to store the URLs
urls = []
```

```python
def extract_urls(menu):
    if 'products' in menu:
        for product in menu['products']:
            if product['productType'] == 'Beverage':
                url = f"/product/{product['productNumber']}/{product['formCode'].lower()}"
                urls.append(url)
    if 'children' in menu:
        for child in menu['children']:
            extract_urls(child)
```

```python
# Extract URLs from the top-level menus
for menu in data['menus']:
    extract_urls(menu)
```

```python
# Display the extracted URLs
for url in urls:
    print(url)
```

```python
# add https://www.starbucks.com/menu/ before and /nutrition after to each URL
nurls = [f"https://www.starbucks.com/menu{url}/nutrition" for url in urls]
```

```python
import pandas as pd
import os
import random
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Replace with the actual path to your chromedriver executable
service = Service("/opt/homebrew/Caskroom/chromedriver/127.0.6533.72/chromedriver-mac-arm64/chromedriver")
driver = webdriver.Chrome(service=service)
```

```python
# Function to scrape data from a single URL
def scrape_product(url):
    driver.get(url)
    try:
        # Wait for elements to load (adjust selectors and wait times as needed)
        name = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'h1.sb-heading.text-bold.color-textWhite.pb2.lg-pb3.productName___NApyx'))
        ).text

        size = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'div.color-textWhiteSoft.auxiliaryProductInfoFont___dnBJo'))
        ).text

        caffeine = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'p#caffeineElement span.text-semibold'))
        ).text

        image_link = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'img.sb-imageFade__imagePositioning.sb-imageFade__show'))
        ).get_attribute("src")

        # Add a delay after scraping (e.g., 5 seconds)
        time.sleep(5)

        return {'name': name, 'size': size, 'caffeine': caffeine, 'link': image_link}
    except Exception as e:
        print(f"Error occurred for URL {url}: {e}")
        return None

```

```python
# Create an empty list to store scraped data
data = []

# Loop through the list of URLs and scrape data
for url in nurls:
    product_data = scrape_product(url)
    data.append(product_data)

# Create a Pandas DataFrame from the scraped data
df = pd.DataFrame(data)

print(df)

# Close the webdriver
driver.quit()
```
