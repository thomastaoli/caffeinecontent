```python
import pandas as pd
import re
```

```python
# Open csv and read it into a pandas dataframe
df = pd.read_csv('caffeine_data.csv')
```

```python
df
```

```python
# Define a function to create the image URL
def create_image_url(link):
    # Extract the part of the link after "/caffeine-content/"
    image_name = link.split('/')[-1]
    # Create the full URL
    return f'https://www.caffeineinformer.com/wp-content/caffeine/{image_name}.jpg'
```

```python
# Apply the function to the 'link' column to create a new 'pic_link' column
df['pic_link'] = df['link'].apply(create_image_url)
```

```python
df
```

```python
# Save the dataframe to a new csv file
df.to_csv('caffeine_data_with_pic_links.csv', index=False)
```
